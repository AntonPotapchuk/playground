```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
eval_simple = np.load('eval_simple_1000.npz')
```

```python
eval_alpha = np.load('eval_alpha_1000.npz')
```

```python
eval_hex = np.load('eval_hex_1000.npz')
```

```python
eval_students = np.load('eval_students_100.npz')
```

```python
reward_means = [np.mean(eval_simple['rewards']), np.mean(eval_alpha['rewards']), np.mean(eval_hex['rewards']), np.mean(eval_students['rewards'])]
reward_stds = [np.std(eval_simple['rewards']), np.std(eval_alpha['rewards']), np.std(eval_hex['rewards']), np.std(eval_students['rewards'])]
plt.bar(range(len(reward_means)), reward_means, yerr=reward_stds)
plt.title("Average reward")
plt.xticks(range(len(reward_means)), ['SimpleAgent', 'AlphaGoZero', 'Hex', 'Students'])
```

```python
length_means = [np.mean(eval_simple['lengths']), np.mean(eval_alpha['lengths']), np.mean(eval_hex['lengths']), np.mean(eval_students['lengths'])]
length_stds = [np.std(eval_simple['lengths']), np.std(eval_alpha['lengths']), np.std(eval_hex['lengths']), np.std(eval_students['lengths'])]
plt.bar(range(len(length_means)), length_means, yerr=length_stds)
plt.title("Average episode length")
plt.xticks(range(len(length_means)), ['SimpleAgent', 'AlphaGoZero', 'Hex', 'Students'])
```

```python
speeds = [eval_simple['elapsed'] / eval_simple['total_timesteps'], eval_alpha['elapsed'] / eval_alpha['total_timesteps'], eval_hex['elapsed'] / eval_hex['total_timesteps'], eval_students['elapsed'] / eval_students['total_timesteps']]
plt.bar(range(len(speeds)), speeds)
plt.title("Time per timestep (s)")
plt.xticks(range(len(speeds)), ['SimpleAgent', 'AlphaGoZero', 'Hex', 'Students'])
```

```python
reward_means = np.array([np.mean(eval_simple['rewards'], axis=0), np.mean(eval_alpha['rewards'], axis=0), np.mean(eval_hex['rewards'], axis=0), np.mean(eval_students['rewards'], axis=0)])
reward_stds = np.array([np.std(eval_simple['rewards'], axis=0), np.std(eval_alpha['rewards'], axis=0), np.std(eval_hex['rewards'], axis=0), np.std(eval_students['rewards'], axis=0)])
plt.bar(np.arange(len(reward_means)), reward_means[:, 0], yerr=reward_stds[:, 0], width=0.2)
plt.bar(np.arange(len(reward_means)) + 0.2, reward_means[:, 1], yerr=reward_stds[:, 1], width=0.2)
plt.bar(np.arange(len(reward_means)) + 0.4, reward_means[:, 2], yerr=reward_stds[:, 2], width=0.2)
plt.bar(np.arange(len(reward_means)) + 0.6, reward_means[:, 3], yerr=reward_stds[:, 3], width=0.2)
plt.title("Average reward per initial position")
plt.xticks(np.arange(len(reward_means)) + 0.3, ['SimpleAgent', 'AlphaGoZero', 'Hex', 'Students'])
plt.legend(['top left', 'bottom left', 'top right', 'bottom right'], bbox_to_anchor=(1.0, 1.0))
```

```python
length_means = np.array([np.mean(eval_simple['lengths'], axis=0), np.mean(eval_alpha['lengths'], axis=0), np.mean(eval_hex['lengths'], axis=0), np.mean(eval_students['lengths'], axis=0)])
length_stds = np.array([np.std(eval_simple['lengths'], axis=0), np.std(eval_alpha['lengths'], axis=0), np.std(eval_hex['lengths'], axis=0), np.std(eval_students['lengths'], axis=0)])
plt.bar(np.arange(len(length_means)), length_means[:, 0], yerr=length_stds[:, 0], width=0.2)
plt.bar(np.arange(len(length_means)) + 0.2, length_means[:, 1], yerr=length_stds[:, 1], width=0.2)
plt.bar(np.arange(len(length_means)) + 0.4, length_means[:, 2], yerr=length_stds[:, 2], width=0.2)
plt.bar(np.arange(len(length_means)) + 0.6, length_means[:, 3], yerr=length_stds[:, 3], width=0.2)
plt.title("Average episode length per initial position")
plt.xticks(np.arange(len(length_means)) + 0.3, ['SimpleAgent', 'AlphaGoZero', 'Hex', 'Students'])
plt.legend(['top left', 'bottom left', 'top right', 'bottom right'], bbox_to_anchor=(1.35, 1.0))
```
